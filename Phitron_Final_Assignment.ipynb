```python
import pandas as pd
import numpy as np

Employees=pd.read_csv('/content/gdrive/MyDrive/final-employee-ds.csv')
print(Employees.head(3))
```

## Answer to the Question no. 01

```python
import numpy as np

TrainingHours=Employees['TrainingHours'].to_numpy()
Convert_into_float=TrainingHours.astype(np.float64)

print(np.mean(Convert_into_float,axis=0))
print(np.std(Convert_into_float,axis=0))
```

## Answer to the Question no. 02

```python
Employees["Salary"] =Employees.groupby("Department")["Salary"].transform(
    lambda x: x.fillna(x.median())
    )
calculate=(Employees["ExperienceYears"]>12) & (Employees["ProjectCount"]>5)
total_salary=Employees.loc[calculate,"Salary"].sum()
print(total_salary)
```

## Answer to the Question no. 03

```python
Filter_data=Employees[(Employees["Department"]=="IT") | (Employees["Department"]=="Finance")]
Data=Filter_data[(Filter_data["Age"]>=30) & (Filter_data["Age"]<=45)]
Data=Data[(Data["PerformanceScore"])>88]
Data=Data.sort_values(by="Salary",ascending=False)
print(Data[["Name","Department","Salary","PerformanceScore"]])
```

## Answer to the Question no. 04

```python
Employees["SalaryPerHour"]=Employees["Salary"]/(Employees["WorkHoursPerWeek"]*4)
Top_5_Employees=Employees.sort_values(by="SalaryPerHour",ascending=False)
print(Top_5_Employees[["Name","SalaryPerHour"]].head(5))
```
